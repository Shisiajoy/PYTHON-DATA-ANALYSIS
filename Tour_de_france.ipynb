```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-04-07/tdf_winners.csv"
tour_de_france_data = pd.read_csv(data_url)
```

```python
tour_de_france_data.info()
```

```python
tour_de_france_data.head(10)
```

```python
tour_de_france_data.isnull().sum()
```

Drop the columns that we're not interested in

```python
tour_de_france_data.drop(columns = ['winner_team', 'born', 'died', 'full_name', 'nickname', 'birth_town', 'birth_country'], inplace=True)
```

```python
tour_de_france_data.shape[1]
```

```python
tour_de_france_data.head()
```

```python
tour_de_france_data.isnull().sum()
```

Remove any row with missing data.

```python
tour_de_france_data.dropna(axis=0, how='any', inplace=True)
```

```python
tour_de_france_data.isnull().sum()
```

Make a list of datetime info using 'start_date' column

```python
tdf_date_time = pd.to_datetime(tour_de_france_data['start_date'])
```

insert new 'year' column using datetime list


```python
tour_de_france_data.insert(2, 'year', tdf_date_time.dt.year)
```

drop start_date columns

```python
tour_de_france_data.drop(columns = 'start_date', inplace=True)
```

change time_margin from hours to minutes & rename column

```python
tour_de_france_data['time_margin'] = tour_de_france_data['time_margin']*60
tour_de_france_data = tour_de_france_data.rename(columns={'time_margin': 'win_margin_min'})
```

```python
tour_de_france_data.head(10)
```

make a final data set excluding the first two rows(the data becomes consistent after 1947)

```python
final_data = tour_de_france_data.iloc[2:,:]
```

```python
final_data.head()
```

```python
final_data.tail()
```

```python
final_data.reset_index(drop=True, inplace=True)
```

```python
final_data.head()
```

Plot heatmap of correlation matrix

```python
import seaborn as sns
# Select only numeric columns
numeric_columns = final_data.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_columns.corr(), annot=True)
```

What relationship is there between a winners age and the number of stages led.

```python
import matplotlib.pyplot as plt

# Calculate the correlation coefficient
correlation_coefficient = final_data['age'].corr(final_data['stages_led'])
print("Correlation Coefficient:", correlation_coefficient)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(final_data['age'], final_data['stages_led'], marker='o', color='blue', alpha=0.6)
plt.xlabel('Winner\'s Age')
plt.ylabel('Number of Stages Led')
plt.title('Relationship between Winner\'s Age and Number of Stages Led')
plt.grid(True)
plt.show()
```

```python
# Sort the DataFrame by the 'year' column in descending order
sorted_data = final_data.sort_values(by='year', ascending=False)

# Find the winner's nationality from the first row
most_recent_winner_nationality = sorted_data.iloc[0]['nationality']

print("Most Recent Winner's Nationality:", most_recent_winner_nationality)
```
